# DATA VISUALIZATION

```python
# DATA VISUALIZATION
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4])
y = x**2 + 3
```

```python
plt.style.available
```

```python
#plt.style.use('dark_background')
plt.style.use('seaborn')
#we can use any design for the graphs
```

```python
print (x, y)
```

```python
#Plot function
plt.plot(x, x**2 +3, label = "Stock Price 1")
plt.plot(x, x**2, label = "Stock Price 2")
plt.title("My first Plot")
plt.legend()   #It shows us the box (left side top)
plt.xlabel("Time stamp")
plt.ylabel("Price")
plt.show()
```

```python
#Scatter function
plt.scatter(x, x**2 +3, label = "Stock Price 1")
plt.scatter(x, x**2, label = "Stock Price 2")
plt.title("My Scatter Plot")
plt.legend()
plt.xlabel("Time stamp")
plt.ylabel("Price")
plt.show()
```

# OPEN CV

```python
#WORKING WITH OPEN CV AND IMAGES
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('shinchan.jpg')
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #to convert bgr to rgb
```

```python
print(img.shape)
```

```python
plt.imshow(rgb_image)   #Showed in bgr form
plt.axis("off") #to remove the grid
plt.show()
```

```python
crop_image = rgb_image[200:1000,100:400,:]  #Croping the image
plt.imshow(crop_image)   #Showed in bgr form
plt.axis("off") #to remove the grid
plt.show()
```

```python
crop_image.shape
```

```python
total_pixels = crop_image.shape[0]*crop_image.shape[1]
```

```python
print(total_pixels)
```

```python
cv2.resize?
```

```python
crop_image = cv2.resize(crop_image,(100,100))
```

```python
plt.imshow(crop_image)   #Showed in bgr form
plt.axis("off") #to remove the grid
plt.title('Shinchan'+str(crop_image.shape))
plt.show()
```
